## 2.Filing Nan with pad/fill(Fill methods Forward) and bfill/backfill(Fill methods Backward)

```python
def f_fill_and_b_fill(df,variable):
    df[variable+' f_fill'] = df[variable].ffill()
    df[variable+' b_fill'] = df[variable].bfill()
```

```python
df2 = pd.read_csv('titanic.csv')
```

```python
df2.head()
```

```python
df2.isnull().sum()
```

```python
f_fill_and_b_fill(df2,'Age')
```

4.statical method

  - mean 

  - median 

  - mode

```python
df4 = pd.read_csv('titanic.csv')
```

```python
df4.head()
```

```python
df4.isnull().sum()
```

```python
mean_age = df4['Age'].mean()
```

```python
mean_age
```

```python
def fill_with_mean(df,variable,mean):
    df[variable+ " mean"]= df[variable].fillna(mean)
```

```python
fill_with_mean(df4,'Age',mean_age)
```

```python
df4
```

```python
df4.isnull().sum()
```

```python
median_age = df4['Age'].median()
median_age
```

```python
def fill_with_median(df,variable,median):
    df[variable+" median"]= df[variable].fillna(median)
```

```python
fill_with_median(df4,'Age',median_age)
```

```python
df4.head()
```

```python
mode_age = df4['Age'].mode()
```

```python
mode_age
```

```python
def fill_with_mode(df,variable,mode):
    df[variable+" mode"]= df[variable].fillna(mode)
```

```python
fill_with_mode(df4,'Age',mode_age)
```

## 6.End of Distribution imputation

```python
df6 = pd.read_csv('titanic.csv')
```

```python
df6.head()
```

```python
def impute_nan(df,variable,median):
    extreme=df.Age.mean()+3*df.Age.std()
    df[variable+"_end_distribution"]=df[variable].fillna(extreme)
    df[variable].fillna(median,inplace=True)
```

```python
impute_nan(df6,'Age',df6.Age.median())
```
