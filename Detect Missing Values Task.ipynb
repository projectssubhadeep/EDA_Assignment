```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.info()
```

```python
titanic.isnull().sum()
```

```python
import missingno as msno
```

## BAR PLOT

```python
msno.bar(titanic)
```

#### The above Bar plot shows that the 'age' and 'deck' columns have lots of missing values. Also 'embarked' and 'embarked_town' have 2 missing values each.
#### The Bar chart does not show which part of the feature has missing value.


## MATRIX PLOT

```python
msno.matrix(titanic)
```

#### In the MATRIX PLOT we can see where the missing values are located.The white lines indicate missing values.

## HEATMAPS

```python
msno.heatmap(titanic)
```

#### HEATMAP shows how missing values are related. 
#### The above heatmap shows only four columns which have missing values.
#### The columns 'embarked' and 'embark_town' correspond to value 1, which means a direct correlation.
#### The columns 'age' and 'deck' have no relation, means the are not related or no correlation.
